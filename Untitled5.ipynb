```python
# Assuming the model is part of the data_dict and is named 'model'
model = data_dict
```

```python
# Extract coefficients
coefficients = model
coeff_series = pd.Series(coefficients, index=feature_names)
```

```python
# Change Pandas display options
pd.options.display.float_format = '{:,.2f}'.format
```

```python
# Display coefficients
print(coeff_series)
```

```python
# Sort coefficients for plotting
coeff_series_sorted = coeff_series.sort_values(ascending=False)

# Create a bar graph of the coefficients
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
coeff_series_sorted.plot(kind='bar')
plt.title('Feature Coefficients (Linear Regression)')
plt.xlabel('Features')
plt.ylabel('Coefficient Value')
plt.show()
```

```python
# Interpret the coefficients
# Three largest positive coefficients
largest_positive = coeff_series_sorted.head(3)
for feature, coef in largest_positive.items():
    print(f"Feature: {feature}, Coefficient: {coef}")
    print(f"Interpretation: A one-unit increase in '{feature}' is associated with an average increase of {coef:.2f} in the target, holding all other features constant.\n")
```

```python
# Three largest negative coefficients
largest_negative = coeff_series_sorted.tail(3)
for feature, coef in largest_negative.items():
    print(f"Feature: {feature}, Coefficient: {coef}")
    print(f"Interpretation: A one-unit increase in '{feature}' is associated with an average decrease of {abs(coef):.2f} in the target, holding all other features constant.\n")
```

```python
# Extract feature importances from the random forest model
importances = model
importances_series = pd.Series(importances, index=feature_names)

# Sort importances for plotting
importances_series_sorted = importances_series.sort_values(ascending=False)

# Create a bar graph of the feature importances
plt.figure(figsize=(10, 6))
importances_series_sorted.plot(kind='bar')
plt.title('Feature Importances (Random Forest)')
plt.xlabel('Features')
plt.ylabel('Importance')
plt.show()

# Top 6 most important features
top_6_features = importances_series_sorted.head(6)
print("Top 6 most important features:")
print(top_6_features)
```

```python
# Check for overlap with top coefficients
top_positive_features = set(largest_positive.index)
top_negative_features = set(largest_negative.index)
top_features = set(top_6_features.index)

overlap_positive = top_positive_features.intersection(top_features)
overlap_negative = top_negative_features.intersection(top_features)

print("Features in top 6 important features that are also in the top 3 largest positive coefficients:", overlap_positive)
print("Features in top 6 important features that are also in the top 3 largest negative coefficients:", overlap_negative)
```

```python
# Apply Shap to Explain the Random Forest Model
### import shap package
import shap
shap.initjs()
```

```python
## sample data
sample_size = 500
X_shap = X_train.sample(sample_size, random_state=321)
y_shap = y_train.loc[X_shap.index]

### inspection
X_shap.isna().sum(), X_shap.info(), y_shap.isnull().sum()
```

```python
## value count data of categorical
X_shap['Central Air'].value_counts(dropna=False), X_shap['Land Contour'].value_counts(dropna=False)
```

```python
# impute for null in Lot Frontage
X_shap['Lot Frontage'].fillna(X_shap['Lot Frontage'].mean(), inplace=True)

X_shap['Lot Frontage'].isna().sum()
```

```python
# replacing object to numeric
X_shap['Central Air'].replace({"Y":1, "N":0}, inplace=True)
X_shap['Land Contour'].replace({ 'Lvl':4, 'Bnk':3, 'HLS':2, 'Low':1 }, inplace=True)
```

```python
X_shap.info()
```

```python
# fit random forest model on shap dataset
from sklearn.ensemble import RandomForestRegressor
shap_model = RandomForestRegressor(n_estimators=100, random_state=321)
shap_model.fit(X_shap, y_shap)
```

```python
# Create a SHAP explainer for the random forest model
explainer = shap.Explainer(shap_model, X_shap)
shap_values = explainer(X_shap)
```

```python
# calculate SHAP values
shap_values = explainer.shap_values(X_shap)
```

```python
# create summary plot
shap.summary_plot(shap_values, X_shap, plot_type='dot')
```

```python
# Interpret top features
## Identify the top 6 features based on the summary plot
print("""# There are:
### #1 Gr Liv Area
### #2 Total Bsmt SF
### #3 Lot Frontage
### #4 Overall Cond
### #5 Lot Area
### #6 TotRms AbvGrd
""")
print()
# Interpret their effects on predictions
print('''# Exaplain as:
###_ #1 Gr Liv Area and #2 Total Bsmt SF ....have higher positive impact values than negative though most values are concentrated on the negative
###_ #3 Lot Frontage and #4 Overall Cond ....tend to have more impact on the negative side
###_ #5 Lot Area and #6 TotRms AbvGrd .......tend to a balanced impact on both direction
''')
```
